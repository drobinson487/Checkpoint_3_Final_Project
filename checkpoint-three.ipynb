```python
#delete unnecessary columns
df_clean.drop(['Unnamed: 0','Size','Founded','Type of ownership','Revenue','Competitors','Headquarters'], axis=1, inplace=True)
```

```python
#delete unwanted str (K,$) in 'Salary Estimate'
salary = df_clean['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary = salary.apply(lambda x: x.replace('K', '').replace('$', '')) 
salary


```

```python
# Seperating salary into min_salary and max_salary columns
df_clean[['min_salary','max_salary']]=salary.str.split('-',1,expand=True)

#Checkpoint to see if code worked
df_clean['min_salary'] 
df_clean['max_salary']

#Replace nan values with 0 in min_salary and max_salary column
df_clean ['min_salary'] = df_clean ['min_salary'].fillna (0)
df_clean ['max_salary'] = df_clean ['max_salary'].fillna (0)

#Changing data type to int to perform calculation
df_clean['min_salary'] = pd.to_numeric(df_clean['min_salary'],errors='coerce')
df_clean = df_clean.replace(np.nan, 0, regex=True)#replace any non numerical values with 0
df_clean['min_salary'] = df_clean['min_salary'].astype(int)#change 'min_salary' to int

df_clean['max_salary'] = pd.to_numeric(df_clean['max_salary'],errors='coerce')
df_clean = df_clean.replace(np.nan, 0, regex=True)#replace any non numerical values with 0
df_clean['max_salary'] = df_clean['max_salary'].astype(int)#change 'max_salary'to int

#Create average salary column
df_clean['average_salary']=(df_clean['min_salary'] + df_clean['max_salary'])/2
df_clean['average_salary']
```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
#Create a function to shorten job titles
def short_titles(title):
    if 'data scientist' in title.lower():
        return 'Data Scientist'
    elif 'data engineer' in title.lower():
        return 'Data Engineer'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'research scientist' in title.lower():
        return 'Research Scientist'
    elif 'laboratory scientist' in title.lower():
        return 'Laboratory Scientist'
    else:
        return 'Other'
    
 #Apply function to job titles column and create a new column 'New_Titles'   
df_clean['New_Titles'] = df_clean['Job Title'].apply(short_titles)
df_clean['New_Titles'].value_counts()
```

```python
#Create a function to rank experience and create a new column 'Entering Level'
def entering_level(title):
    if 'senior' in title.lower() or 'sr' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower() or 'director'in title.lower() or 'manager' in title.lower():
        return 'Senior Level'
    elif 'junior' in title.lower() or 'jr' in title.lower() or 'jr.' in title.lower():
        return 'Junior Level'
    else:
        return 'Entry Level'
    
#Apply function to job titles column and create a new column 'Entering_level' 
df_clean['Entering_Level'] = df_clean['Job Title'].apply(entering_level)
df_clean['Entering_Level'].value_counts()
```

```python
#Create two new columns 'City' and 'State' from the 'location' column
df_clean[['City','State']]=df_clean['Location'].str.split(',',1,expand=True)

df_clean['City']
```
